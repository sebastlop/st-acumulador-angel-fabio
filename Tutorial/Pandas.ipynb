## Creación de un DataFrame

Una forma común de crear un DataFrame es mediante el uso de diccionarios en Python. En este contexto, las claves del diccionario representan las columnas del DataFrame, mientras que los valores asociados a estas claves representan las filas, siendo estos valores una lista o un arreglo de datos correspondientes a cada columna.

Este enfoque proporciona una manera conveniente de estructurar los datos para su posterior conversión en un DataFrame de pandas. Una vez creado, el DataFrame puede ser utilizado para realizar análisis de datos, manipulaciones y visualizaciones.

```python
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)
df
```
